1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss
nominal, ordinal, interval, and ratio scales.
 - Qualitative (categorical):

    Nominal: labels with no ranking — e.g., blood types A/B/AB/O.
• Ordinal: ordered labels, size of gaps unknown — e.g., survey choices “poor, fair, good, excellent.”

 - Quantitative (numerical):

    Interval: numbers with equal spacing but no true zero — e.g., temperature in °C (0 °C ≠ “no heat”).

     Ratio: numbers with equal spacing and a meaningful zero — e.g., weight in kg (0 kg = no weight).-

2. What are the measures of central tendency, and when should you use each? Discuss the mean, median,
and mode with examples and situations where each is appropriate.
 - Measures of central tendency describe the center of a data set:
 - Mean (average): Sum of all values ÷ number of values.
Best for symmetric data without outliers.
Example: Average marks in a class.

 - Median (middle value): Middle number when data is ordered.
• Best for skewed data or with outliers.
• Example: Median income in a population.

 - Mode (most frequent value): Value that appears most often.
• Best for categorical or bimodal data.
• Example: Most common shoe size sold.

3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?
 - Dispersion refers to the extent to which data values spread out from the center (mean).
 - Variance measures the average squared distance from the mean.
 - Standard deviation is the square root of variance, showing spread in original units.

4. What is a box plot, and what can it tell you about the distribution of data?
 - A box plot is a visual summary of data distribution using five values: minimum, Q1, median, Q3, and maximum.
 - It shows the spread, center, and skewness of data and helps detect outliers easily.

5. Discuss the role of random sampling in making inferences about populations
 - Random sampling ensures every member of a population has an equal chance of being selected.
 - It helps avoid bias, making the sample more representative, which allows for accurate inferences and generalizations about the entire population.

6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?
 - Skewness measures the asymmetry of a data distribution.
 - Positive skew: Tail on the right; mean > median.
 - Negative skew: Tail on the left; mean < median.
 - Symmetrical: Mean ≈ median.
 - Skewness affects how we interpret the center and spread of data, influencing which measure of central tendency to use.

7. What is the interquartile range (IQR), and how is it used to detect outliers?
 - The interquartile range (IQR) is the difference between the third quartile (Q3) and the first quartile (Q1), i.e., IQR = Q3 - Q1.
 - It measures the spread of the middle 50% of data. Outliers are typically detected by identifying values that fall below Q1 - 1.5×IQR or above Q3 + 1.5×IQR.

8. Discuss the conditions under which the binomial distribution is used.
 - The binomial distribution is used when:
There are a fixed number of trials (n),
Each trial has only two possible outcomes (success or failure),
The probability of success (p) is constant for each trial, and
The trials are independent of each other.

9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).
 - The normal distribution is symmetric, bell-shaped, with the mean = median = mode. It is defined by its mean and standard deviation.
The empirical rule states that:
68% of data lies within 1 standard deviation,
95% within 2, and 99.7% within 3 standard deviations from the mean.

11. Explain what a random variable is and differentiate between discrete and continuous random variables.
 - A random variable is a numerical outcome of a random process.
 - A discrete random variable takes countable values (e.g., number of students).
 - A continuous random variable takes any value within a range (e.g., height or weight).
